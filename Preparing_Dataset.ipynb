### Delete the column ‚Äòcustomerid‚Äô from the dataset.

```python
del df['customerID']
```

### Convert the column ‚ÄòTotalCharges‚Äô to a numeric format, coercing errors if necessary.

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

### Change all column names to lowercase.

```python
df = df.rename(str.lower, axis='columns')
df.columns
```

### For string values change spaces to an underscore symbol.

```python
df= df.applymap(lambda s:s.replace(' ', '_') if type(s) == str else s)
```

### Convert the column ‚Äòchurn‚Äô to integer format (no - 0, yes - 1).

```python
df= df.applymap(lambda s: s.replace('No', '0') if type(s) == str else s)
df= df.applymap(lambda s: s.replace('Yes', '1') if type(s) == str else s)
```
